**chi2- test
**

```python
import scipy.stats as stats
```

```python
import seaborn as sns
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np
data = sns.load_dataset("tips")
```

```python
data.head()
```

```python
data_table = pd.crosstab(data["sex"], data["smoker"])
data_table
```

```python
observed_valus = data_table.values
observed_valus
```

```python
val = stats.chi2_contingency(data_table)
val
```

```python
Expected_values = val[3]
Expected_values
```

```python
no_of_rows = len(data_table.iloc[0:2,0])
no_of_columns = len(data_table.iloc[0,0:2])
ddof = (no_of_rows - 1) * (no_of_columns - 1)
print("Degree of Freedom: ", ddof)
alpha = 0.05
```

```python
from scipy.stats import chi2
chi_square = sum([(o - e) ** 2. / e for o, e in zip(observed_valus, Expected_values)])
chi_square_statistic = chi_square[0] + chi_square[1]
```

```python
print("chi-square statistic: ", chi_square_statistic)
```

```python

critical_value = chi2.ppf(q = 1 - alpha, df = ddof)
print("Critical Value: ", critical_value)
```

```python
p_value = 1 - chi2.cdf(x = chi_square_statistic, df = ddof)
print("p-value: ", p_value)
print("Significance Level: ", alpha)
print("Degree of Freedom: ", ddof)
```

```python
if chi_square_statistic >= critical_value:
    print("Reject H0, There is a relationship between 2 categorical variables")
else:
    print("Retain H0, There is no relationship between 2 categorical variables")
```

```python
if p_value <= alpha:
    print("Reject H0, There is a relationship between 2 categorical variables")
else:
    print("Retain H0, There is no relationship between 2 categorical variables")
```

```python
import seaborn as sns

df1= sns.load_dataset('iris')
```

```python
df1.head()
```

```python
df_anova = df1[['petal_width','species']]
```

```python
grps = pd.unique(df_anova.species.values)


grps
```

```python
d_data = {grp:df_anova['petal_width'][df_anova.species == grp] for grp in grps}
```

```python
d_data
```

```python
F, p = stats.f_oneway(d_data['setosa'], d_data['versicolor'], d_data['virginica'])
```

```python
p
```

```python

F
```

```python
if p<0.05:
    print("reject null hypothesis")
else:
    print("accept null hypothesis")
```
